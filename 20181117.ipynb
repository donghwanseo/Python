# Load Dataset
* 데이터는 SAS데이터셋을 이용한다.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_sas('grip_str.sas7bdat').dropna() 
```

```python
#자료가 제대로 읽어졌는지 확인한다.
df.head()
```

```python
# 나이를 구간으로 나누기 위해 Label을 만든다.
labels = ["{0} - {1}".format(i, i + 10) for i in range(19, 80, 10)]
print(labels)
```

```python
df['age'] = df.age.astype('int64')
df['age'] = pd.cut(df.age, range(19, 90, 10), right=False, labels=labels)
df.head()
```

```python
df['sex']=df.sex.astype('int64')
df=pd.concat([df,pd.get_dummies(df.sex, prefix='sex')],axis=1)
df.head()
```

```python
df=pd.concat([df,pd.get_dummies(df.age,prefix='age')],axis=1)
df.head()
```

```python
df=pd.concat([df,pd.get_dummies(df.ho_incm, prefix='ho_incm')],axis=1)
df.head()
```

```python
df=pd.concat([df,pd.get_dummies(df.edu, prefix='edu')],axis=1)
df.head()
```

```python
df=pd.concat([df,pd.get_dummies(df.occp, prefix='occp')],axis=1)
df.head()
```

```python
df=pd.concat([df,pd.get_dummies(df.marri_1, prefix='marri_1')],axis=1)
df.head()
```

```python
df=pd.concat([df,pd.get_dummies(df.sm_presnt, prefix='sm_presnt')],axis=1)
df.head()
```

```python
df=pd.concat([df,pd.get_dummies(df.dr_high, prefix='dr_hight')],axis=1)
df.head()
```

```python
df=pd.concat([df,pd.get_dummies(df.pa_aerobic, prefix='pa_aerobic')],axis=1)
df.head()
```

```python
df=pd.concat([df,pd.get_dummies(df.pa_walk, prefix='pa_walk')],axis=1)
df.head()
```

```python
df=pd.concat([df,pd.get_dummies(df.pa_muscle, prefix='pa_muscle')],axis=1)
df.head()
```

```python
df=pd.concat([df,pd.get_dummies(df.HE_HP, prefix='HE_HP')],axis=1)
df.head()
```

```python
df=pd.concat([df,pd.get_dummies(df.HE_DM, prefix='HE_DM')],axis=1)
df.head()
```

```python
df=pd.concat([df,pd.get_dummies(df.HE_HCHOL, prefix='HE_HCHOL')],axis=1)
df.head()
```

```python
df=pd.concat([df,pd.get_dummies(df.GS_use, prefix='GS_use')],axis=1)
df.head()
```

```python
df=df.drop(['sex','age','ho_incm','edu','occp','marri_1','sm_presnt','dr_high','pa_aerobic','pa_walk','pa_muscle','HE_HP','HE_DM','HE_HCHOL','GS_use'],axis=1)
```

# Feature, Target 설정
* Target 컬럼 생성 - 왼손 악력과 오른손 악력 중 최대값을 타겟으로 한다.

```python
target = df[['max_grip_rt', 'max_grip_lt']].max(axis = 1)
features = df.drop(columns = ['ID', 'max_grip_rt', 'max_grip_lt'], axis = 1)
```

```python
# Train데이터와 Test를 나눈다.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=0)
```

#MinMaxScaler
class sklearn.preprocessing.MinMaxScaler(feature_range=(0, 1), copy=True)

Document:
https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train) 
X_test_scaled = scaler.transform(X_test) 
```

#Ridge
class sklearn.linear_model.Ridge(alpha=1.0, fit_intercept=True, normalize=False, copy_X=True, max_iter=None, tol=0.001, solver=’auto’, random_state=None)

* 규제 방식 : L2 규제

Documentation :  https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1)
```

```python
ridge.fit(X_train_scaled, y_train)
```

```python
ridge.score(X_test_scaled, y_test)
```

alpha에 Default Value를 넣고 Ridge회귀를 돌린 결과 Scoresms 0.747을 얻을 수 있었다.
이제 alpha 값을 변경시켜 score를 증가시키려고 한다.

# KFold, GridSearchCV
class sklearn.model_selection.KFold(n_splits=’warn’, shuffle=False, random_state=None)

class sklearn.model_selection.GridSearchCV(estimator, param_grid, scoring=None, fit_params=None, n_jobs=None, iid=’warn’, refit=True, cv=’warn’, verbose=0, pre_dispatch=‘2*n_jobs’, error_score=’raise-deprecating’, return_train_score=’warn’)

* 회귀이므로 KFold이용

Document:
https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html

Document:
https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits=10, shuffle=True)

param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}

from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(ridge, param_grid = param_grid, cv=kfold)

# 학습
grid_result = grid.fit(X_train_scaled, y_train)
```

```python
print(grid_result.best_score_)
```

```python
grid.best_params_
```

alpha가 0.1일때 모형의 성능이 가장 좋은 것으로 나왔다.
이에 따라 좀더 세분화해서 다시 돌려보기로 한다.

```python
import numpy as np
alpha = np.arange(0.05, 0.3, 0.01)

print(alpha)
```

```python
param_grid = {'alpha': alpha}

grid = GridSearchCV(ridge, param_grid = param_grid, cv=kfold)

# 학습
grid_result = grid.fit(X_train_scaled, y_train)
```

```python
print(grid_result.best_score_)
```

```python
grid.best_params_
```

kfold의 shuffle 매개변수가 True이므로 돌릴때마다 다른 알파값이 최적의 매개변수로 나온다.


그러나 score차이가 크지 않으므로 기존의 alpha가 0.1일때 최적이라고 판단한다.

```python
# 평가
grid.score(X_test_scaled, y_test)
```

최적의 매개변수를 찾아 Score를 구했다.


그러나 문제가 생겼다.

**기존에 디폴트 값으로 alpha를 1로 넣었었는데 오히려 디폴트 모형이 Score가 더 좋게 나왔다.**(혼돈)



# RandomForestRegressor

class sklearn.ensemble.RandomForestRegressor(n_estimators=’warn’, criterion=’mse’, max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features=’auto’, max_leaf_nodes=None, min_impurity_decrease=0.0, min_impurity_split=None, bootstrap=True, oob_score=False, n_jobs=None, random_state=None, verbose=0, warm_start=False)

Document : https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html#sklearn.ensemble.RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=10 , random_state=0)
```

```python
model.fit(X_train_scaled, y_train)
```

```python
model.score(X_test_scaled, y_test)
```

매개변수를 디폴트로 돌렸을때 Score는 0.71을 얻었다. 이를 GridSearch를 이용하여 향상시키고자 한다.

# GridSearchCV

```python
kfold = KFold(n_splits=5, shuffle=True) # 소요시간이 길어 n_splits을 10에서 5로 줄였다

estimators = np.arange(10, 100, 10)

param_grid = {'n_estimators': estimators}

grid = GridSearchCV(model, param_grid = param_grid, cv=kfold)

# 학습
grid_result = grid.fit(X_train_scaled, y_train)
```

```python
print(grid_result.best_score_)
```

```python
grid.best_params_
```

```python
# 평가
grid.score(X_test_scaled, y_test)
```

GridSearch를 이용하여 Score를 0.03정도 증가 시킬 수 있었다.
